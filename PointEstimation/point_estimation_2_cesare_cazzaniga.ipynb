```python
import numpy as np

# 1. MLE
def lnL(xx,yy,a,b):
    """function returning -2*ln(L) """ ##up to a constant
    return len(xx)*np.log(2*np.pi) + np.inner(yy - b*xx - a ,yy - b*xx - a )

def sl_MLE(xx,yy):
    """compute the MLE for the straight line (sl) parameters"""
    b = (np.mean(xx*yy)-np.mean(xx)*np.mean(yy))/(np.mean(np.power(xx,2)) - np.power(np.mean(xx),2)) 
    a = np.mean(yy) - b*np.mean(xx)
    return (a,b)

def sl_MLE_var(xx,yy):
    """compute the covariance matrix of the MLE"""
    V11 = (np.mean(np.power(xx,2)))/(len(xx)*(np.mean(np.power(xx,2)) - np.power(np.mean(xx),2)))   #for a
    V12 = - (np.mean(xx))/(len(xx)*(np.mean(np.power(xx,2)) - np.power(np.mean(xx),2)))
    V21 = V12
    V22 = 1./(len(xx)*(np.mean(np.power(xx,2)) - np.power(np.mean(xx),2)))                        #for b
    return ((V11,V12),(V21,V22))

# position of the detector hits
xx12 = np.array([10,11,12,20,21,22])
yy12 = np.array([4.0,3.8,3.6,5.2,4.9,4.8])

# sub-arrays with only station 1 or only station 2
xx1 = xx12[[0,1,2]]
xx2 = xx12[[3,4,5]]
yy1 = yy12[[0,1,2]]
yy2 = yy12[[3,4,5]]

print ("first station only: V(MLE) = ", sl_MLE_var(xx1,yy1))
print ("second station only: V(MLE) = ", sl_MLE_var(xx2,yy2))
print ("both stations: V(MLE) = ", sl_MLE_var(xx12,yy12))
mle12 = sl_MLE(xx12,yy12) 
print ("MLE for both stations: ", mle12)

mle1 = sl_MLE(xx1,yy1)
mle2 = sl_MLE(xx2,yy2)

print ("MLE for 1st station: ", mle1)
print ("MLE for 2nd station: ", mle2)
```

Parameters estimators from MLE
$$ \hat{b} = \frac{\bar{xy} - \bar{x}\bar{y}}{\bar{x^2} - \bar{x}^2} $$
$$ \hat{a} = \bar{y} -  \hat{b}\bar{x}$$ 
Variances:
$$ \sigma^2_{a} = \frac{\sigma^2 \bar{x^2}}{N(\bar{x^2} - \bar{x}^2)}  $$
$$ \sigma^2_{b} = \frac{\sigma^2 \bar{x^2}}{(\bar{x^2} - \bar{x}^2)}$$
Covariances:
$$ cov(\hat{a},\hat{b}) = -\frac{\sigma^2 \bar{x}}{N(\bar{x^2} - \bar{x}^2)}$$

```python
# 2. draw -2 ln L

# 'wrappers' for the optimization (which happens with respect to the first parameter)
def lnL_args_ab(x,*args):
    """for minimizing with respect to both a and b"""
    a = x[0]
    b = x[1]
    xx = args[0]
    yy = args[1]
    return lnL(xx,yy,a,b)

def lnL_args_a(x,*args):
    """for minimizing with respect to only b"""
    a = args[2]
    b = x[0]
    xx = args[0]
    yy = args[1]
    return lnL(xx,yy,a,b)

# -2ln(L) values for both stations combined
lh_vals12 = []
# profile -2ln(L) values for both stations combined
proflh_vals12 = []
# -2ln(L) values for the first station only
lh_vals1 = []
# profile -2ln(L) values for the first station only
proflh_vals1 = []
# -2ln(L) values for the second station only
lh_vals2 = []
# profile -2ln(L) values for the second station only
proflh_vals2 = []

# minimum -2ln(L) for both stations combined
minlh12 = lnL(xx12,yy12,sl_MLE(xx12,yy12)[0],sl_MLE(xx12,yy12)[1])
# minimum -2ln(L) for the first station only
minlh1 = lnL(xx1,yy1,sl_MLE(xx1,yy1)[0],sl_MLE(xx1,yy1)[1])
# minimum -2ln(L) for the second station only 
minlh2 =  lnL(xx2,yy2,sl_MLE(xx2,yy2)[0],sl_MLE(xx2,yy2)[1])

print ("minlh12 = ", minlh12 )
print ("minlh1 = ", minlh1 )
print ("minlh2 = ", minlh2 )

# array with the a values that we are going to scan
a_vals = np.linspace(0,15,150)


for ai in a_vals:
    # 12
    lh_vals12.append(lnL(xx12,yy12,ai, sl_MLE(xx12,yy12)[1]) -  minlh12) #TODO
    # LH minimisation: use https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fmin.html
    bhathat = optimize.fmin(lnL_args_a, sl_MLE(xx12,yy12)[1], (xx12,yy12,ai))    
    proflh_vals12.append(lnL(xx12, yy12, ai, bhathat[0]) -  minlh12) 
    # 1
    lh_vals1.append(lnL(xx1,yy1, ai, sl_MLE(xx1,yy1)[1]) -  minlh1 ) 
    bhathat = optimize.fmin(lnL_args_a, sl_MLE(xx1,yy1)[1], (xx1,yy1,ai))
    proflh_vals1.append(lnL(xx1, yy1, ai, bhathat[0]) -  minlh1) 
    # 2
    lh_vals2.append(lnL(xx2,yy2,ai, sl_MLE(xx2,yy2)[1]) -  minlh2 ) 
    bhathat = optimize.fmin(lnL_args_a, sl_MLE(xx2,yy2)[1], (xx2,yy2,ai)) 
    proflh_vals2.append(lnL(xx2, yy2, ai, bhathat[0]) -  minlh2) 



# plotting
fix, ax = plt.subplots()
ax.plot(a_vals, proflh_vals1, label='$L_{prof,1} (a)$', linestyle='-', color='r')
ax.plot(a_vals, lh_vals1, label='$L_{1} (a,\hat{b})$', linestyle='--', color='r')
ax.plot(a_vals, proflh_vals2, label='$L_{prof,2} (a)$', linestyle='-', color='b')
ax.plot(a_vals, lh_vals2, label='$L_{2} (a,\hat{b})$', linestyle='--', color='b')
ax.plot(a_vals, proflh_vals12, label='$L_{prof,1+2} (a)$', linestyle='-', color='black')
ax.plot(a_vals, lh_vals12, label='$L_{1+2} (a,\hat{b})$', linestyle='--', color='black')
ax.set_ybound(0,6)
ax.set_xlabel('a')
ax.set_ylabel('$-2 \Delta \ln(\mathcal{L})$')
ax.legend()
plt.show()
```

```python
# 1. MLE
def lnL(xx,yy,a,b):
    """function returning -2*ln(L) """ ##up to a constant
    return len(xx)*np.log(2*np.pi) + np.inner(yy - b*xx - a ,yy - b*xx - a )

def sl_MLE(xx,yy):
    """compute the MLE for the straight line (sl) parameters"""
    b = (np.mean(xx*yy)-np.mean(xx)*np.mean(yy))/(np.mean(np.power(xx,2)) - np.power(np.mean(xx),2)) 
    a = np.mean(yy) - b*np.mean(xx)
    return (a,b)

def sl_MLE_var(xx,yy):
    """compute the covariance matrix of the MLE"""
    V11 = (np.mean(np.power(xx,2)))/(len(xx)*(np.mean(np.power(xx,2)) - np.power(np.mean(xx),2)))   #for a
    V12 = - (np.mean(xx))/(len(xx)*(np.mean(np.power(xx,2)) - np.power(np.mean(xx),2)))
    V21 = V12
    V22 = 1./(len(xx)*(np.mean(np.power(xx,2)) - np.power(np.mean(xx),2)))                        #for b
    return ((V11,V12),(V21,V22))

# position of the detector hits
xx12 = np.array([10,11,12,20,21,22])
yy12 = np.array([4.0,3.8,3.6,5.2,4.9,4.8])

# sub-arrays with only station 1 or only station 2
xx1 = xx12[[0,1,2]]
xx2 = xx12[[3,4,5]]
yy1 = yy12[[0,1,2]]
yy2 = yy12[[3,4,5]]

cov1 = sl_MLE_var(xx1,yy1)
cov2 = sl_MLE_var(xx2,yy2)
cov12 = sl_MLE_var(xx12,yy12)

print ("first station only: V(MLE) = ", cov1)
print ("second station only: V(MLE) = ", cov2)
print ("both stations: V(MLE) = ",cov12)
mle12 = sl_MLE(xx12,yy12) 
print ("MLE for both stations: ", mle12)

mle1 = sl_MLE(xx1,yy1)
mle2 = sl_MLE(xx2,yy2)

print ("MLE for 1st station: ", mle1)
print ("MLE for 2nd station: ", mle2)


#BLUE method 

cov1 = np.matrix(cov1)
cov2 = np.matrix(cov2)

cova12 = np.matrix([[np.sqrt(cov1[0,0])*np.sqrt(cov1[0,0]), np.sqrt(cov1[0,0])*np.sqrt(cov2[0,0])], [np.sqrt(cov1[0,0])*np.sqrt(cov2[0,0]),np.sqrt(cov2[0,0])*np.sqrt(cov2[0,0])]])

invcova12 = np.linalg.inv(cova12)

print("combined cov matrix MLE = ", invcova12)

invcova12arr = np.array(invcova12)

sum12 = np.sum(invcova12arr)

print ("sum invcov: ", sum12)

w1 = (invcova12[0,0] + invcova12[0,1])/(sum12)
w2 = (invcova12[1,0] + invcova12[1,1])/(sum12)

print ("w1: ", w1)
print ("w2: ", w2)

xBLUE = w1*sl_MLE(xx1,yy1)[0] + w2*sl_MLE(xx2,yy2)[0]

print ("ahat1", sl_MLE(xx1,yy1)[0])
print ("ahat2", sl_MLE(xx2,yy2)[0])

print ("aBLUE:", xBLUE)

```
